```python
import os
import json
import xml.etree.ElementTree as ET
import numpy as np
import shutil
```

```python
def xml_reader(xml):
    dataset = []
    tree = ET.parse(xml)
    root = tree.getroot()
    for img in root[2:]:
        data = dict()
        iname = img.attrib['name'].split('/')[3]
        iname_arr = iname[:-4].split('_')
        if iname_arr[7] == '45':
            iname_arr[7] = '30'
        elif iname_arr[7] == '90':
            iname_arr[7] = '60'
        elif iname_arr[7] == 'minus45':
            iname_arr[7] = 'minus30'
        elif iname_arr[7] == 'minus90':
            iname_arr[7] = 'minus60'
        data['filename'] = f"sample_data\\{iname_arr[0]}\\{'heat_img' if 'h' in iname_arr else 'rgb'}\\{'_'.join(iname_arr)}.png"
        data['height'] = img.attrib['height']
        data['width'] = img.attrib['width']
        data['distance'] = iname_arr[3]
        data['area_temper'] = iname_arr[5]
        data['person_temper'] = iname_arr[4]
        data['humidity'] = iname_arr[6]
        data['gender'] = iname_arr[0][0]
        data['age'] = iname_arr[1]
        data['degree'] = [iname_arr[7], iname_arr[8]]
        data['annotation'] = []
        labels = {}
        for label in img:
            labels[label.attrib['label']] = label
        face = dict()
        face['label'] = labels['face'].attrib['label']
        face['box'] = [round(float(labels['face'].attrib['xtl'])), round(float(labels['face'].attrib['ytl'])), round(float(labels['face'].attrib['xbr'])),round(float(labels['face'].attrib['ybr']))]
        face['keypoint'] = dict()
        face['keypoint']['forehead'] = [round(float(labels['forehead'].attrib['points'].split(',')[0])),round(float(labels['forehead'].attrib['points'].split(',')[1])), 1 if labels['forehead'].attrib['occluded'] == '0' else 0]
        face['keypoint']['glabella'] = [round(float(labels['glabella'].attrib['points'].split(',')[0])),round(float(labels['glabella'].attrib['points'].split(',')[1])), 1 if labels['glabella'].attrib['occluded'] == '0' else 0]
        face['keypoint']['l_eye'] = [round(float(labels['l_eye'].attrib['points'].split(',')[0])),round(float(labels['l_eye'].attrib['points'].split(',')[1])), 1 if labels['l_eye'].attrib['occluded'] == '0' else 0]
        face['keypoint']['r_eye'] = [round(float(labels['r_eye'].attrib['points'].split(',')[0])),round(float(labels['r_eye'].attrib['points'].split(',')[1])), 1 if labels['r_eye'].attrib['occluded'] == '0' else 0]
        face['keypoint']['nose'] = [round(float(labels['nose'].attrib['points'].split(',')[0])),round(float(labels['nose'].attrib['points'].split(',')[1])), 1 if labels['nose'].attrib['occluded'] == '0' else 0]
        if 'h' not in iname:
            face['temper'] = dict()
            np_data = np.load(f"sample_data\\{iname_arr[0]}\\heat\\{'_'.join(iname_arr)}.npy")
            for h_img in root[2:]:
                if h_img.attrib['name'] == f"sample_data/{iname_arr[0]}/heat_img/{iname[:-4]}_h.png":
                    h_lables = dict()
                    for h_label in h_img:
                        if h_label.attrib['label'] in ['r_eye', 'forehead', 'nose', 'glabella', 'l_eye']:
                            h_lables[h_label.attrib['label']] = h_label
                    face['temper']['forehead'] = np_data[round(float(h_lables['forehead'].attrib['points'].split(',')[1]))][round(float(h_lables['forehead'].attrib['points'].split(',')[0]))]
                    face['temper']['glabella'] = np_data[round(float(h_lables['glabella'].attrib['points'].split(',')[1]))][round(float(h_lables['glabella'].attrib['points'].split(',')[0]))]
                    face['temper']['l_eye'] = np_data[round(float(h_lables['l_eye'].attrib['points'].split(',')[1]))][round(float(h_lables['l_eye'].attrib['points'].split(',')[0]))]
                    face['temper']['r_eye'] = np_data[round(float(h_lables['r_eye'].attrib['points'].split(',')[1]))][round(float(h_lables['r_eye'].attrib['points'].split(',')[0]))]
                    face['temper']['nose'] = np_data[round(float(h_lables['nose'].attrib['points'].split(',')[1]))][round(float(h_lables['nose'].attrib['points'].split(',')[0]))]
        data['annotation'].append(face)
        for label in labels:
            if label not in ['face', 'r_eye', 'forehead', 'nose', 'glabella', 'l_eye']:
                label = labels[label]
                etc = dict()
                etc['label'] = label.attrib['label']
                etc['box'] = [round(float(label.attrib['xtl'])), round(float(label.attrib['ytl'])), round(float(label.attrib['xbr'])), round(float(label.attrib['ybr']))]
                if 'h' not in iname:
                    if label.attrib['label'] == 'black1':
                        if iname_arr[0] in ['man1','man2','man3','man4','man5','woman2','woman5']:
                            if iname_arr[3] == '1.0':
                                etc['depth'] = 1.234
                            elif iname_arr[3] == '1.5':
                                etc['depth'] = 1.713
                            elif iname_arr[3] == '2.0':
                                etc['depth'] = 2.167
                            elif iname_arr[3] == '2.5':
                                etc['depth'] = 2.735
                            elif iname_arr[3] == '3.0':
                                etc['depth'] = 3.227
                        else:
                            if iname_arr[3] == '1.0':
                                etc['depth'] = 1.248
                            elif iname_arr[3] == '1.5':
                                etc['depth'] = 1.731
                            elif iname_arr[3] == '2.0':
                                etc['depth'] = 2.166
                            elif iname_arr[3] == '2.5':
                                etc['depth'] = 2.730
                            elif iname_arr[3] == '3.0':
                                etc['depth'] = 3.229
                        for h_img in root[2:]:
                            if h_img.attrib['name'] == f"sample_data/{iname_arr[0]}/heat_img/{iname[:-4]}_h.png":
                                for h_label in h_img:
                                    if h_label.attrib['label'] == 'black1':
                                        xmin,ymin,xmax,ymax = round(float(h_label.attrib['xtl'])),round(float(h_label.attrib['ytl'])),round(float(h_label.attrib['xbr'])),round(float(h_label.attrib['ybr']))
                                        xcenter = round((xmax - xmin)/2 + xmin)
                                        ycenter = round((ymax - ymin)/2 + ymin)
                                        np_data = np.load(f"sample_data\\{iname_arr[0]}\\heat\\{'_'.join(iname_arr)}.npy")
                                        np_data_center = [np_data[ycenter-1][xcenter-1], np_data[ycenter-1][xcenter], np_data[ycenter-1][xcenter+1],
                                                          np_data[ycenter][xcenter-1], np_data[ycenter][xcenter], np_data[ycenter][xcenter+1],
                                                          np_data[ycenter+1][xcenter-1], np_data[ycenter+1][xcenter], np_data[ycenter+1][xcenter+1]]
                                        np_data_mean = sum(np_data_center)/len(np_data_center)
                                        etc['temper'] = np_data_mean
                    elif label.attrib['label'] == 'black2':
                        if iname_arr[0] in ['man1','man2','man3','man4','man5','woman2','woman5']:
                            if iname_arr[3] == '1.0':
                                etc['depth'] = 1.218
                            elif iname_arr[3] == '1.5':
                                etc['depth'] = 1.696
                            elif iname_arr[3] == '2.0':
                                etc['depth'] = 2.147
                            elif iname_arr[3] == '2.5':
                                etc['depth'] = 2.716
                            elif iname_arr[3] == '3.0':
                                etc['depth'] = 3.207
                        else:
                            if iname_arr[3] == '1.0':
                                etc['depth'] = 1.208
                            elif iname_arr[3] == '1.5':
                                etc['depth'] = 1.700
                            elif iname_arr[3] == '2.0':
                                etc['depth'] = 2.134
                            elif iname_arr[3] == '2.5':
                                etc['depth'] = 2.701
                            elif iname_arr[3] == '3.0':
                                etc['depth'] = 3.203
                        for h_img in root[2:]:
                            if h_img.attrib['name'] == f"sample_data/{iname_arr[0]}/heat_img/{iname[:-4]}_h.png":
                                for h_label in h_img:
                                    if h_label.attrib['label'] == 'black2':
                                        xmin,ymin,xmax,ymax = round(float(h_label.attrib['xtl'])),round(float(h_label.attrib['ytl'])),round(float(h_label.attrib['xbr'])),round(float(h_label.attrib['ybr']))
                                        xcenter = round((xmax - xmin)/2 + xmin)
                                        ycenter = round((ymax - ymin)/2 + ymin)
                                        np_data = np.load(f"sample_data\\{iname_arr[0]}\\heat\\{'_'.join(iname_arr)}.npy")
                                        np_data_center = [np_data[ycenter-1][xcenter-1], np_data[ycenter-1][xcenter], np_data[ycenter-1][xcenter+1],
                                                          np_data[ycenter][xcenter-1], np_data[ycenter][xcenter], np_data[ycenter][xcenter+1],
                                                          np_data[ycenter+1][xcenter-1], np_data[ycenter+1][xcenter], np_data[ycenter+1][xcenter+1]]
                                        np_data_mean = sum(np_data_center)/len(np_data_center)
                                        etc['temper'] = np_data_mean
                data['annotation'].append(etc)
        dataset.append(data)
    return dataset
```

```python
dataset = xml_reader('annotations.xml')
with open('sample_data.json', 'w', encoding='utf-8') as make_file:
    json.dump(dataset, make_file, indent="\t")
```
